Task 2 – Time Series Analysis & Decomposition

```python
#Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
```

```python
#Load Dataset
#Change the path if your CSV is in another folder

sales_df = pd.read_csv("train excel/train.csv")
```

```python
#Convert Dates (Works for every Superstore dataset)

sales_df["Order Date"] = pd.to_datetime(
    sales_df["Order Date"],
    errors="coerce",
    dayfirst=False
)

sales_df["Ship Date"] = pd.to_datetime(
    sales_df["Ship Date"],
    errors="coerce",
    dayfirst=False
)

print(sales_df.dtypes)
```

```python
#Remove Invalid Dates

sales_df = sales_df.dropna(subset=["Order Date"])
```

```python
#Create Monthly Sales

monthly_sales = (
    sales_df.set_index("Order Date")
            .resample("ME")
            .agg({"Sales":"sum"})
)

monthly_sales.head()
```

```python
#Plot Monthly Sales Trend

plt.figure(figsize=(15,6))

plt.plot(monthly_sales.index,
         monthly_sales["Sales"],
         marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.grid(True)

plt.savefig(
    "charts/monthly_sales_trend.png",
    bbox_inches="tight"
)

plt.show()
```

Observation:

Monthly sales fluctuate over time but generally show an increasing trend.
Noticeable peaks occur towards the end of several years, indicating seasonal demand.
The data suggests that sales are influenced by yearly shopping patterns.

```python
#Time Series Decomposition

decomposition = seasonal_decompose(
    monthly_sales["Sales"],
    model="additive",
    period=12
)

fig = decomposition.plot()

fig.set_size_inches(15,10)

plt.savefig(
    "charts/time_series_decomposition.png",
    bbox_inches="tight"
)

plt.show()
```

Observations:

• The trend component shows overall business growth over time.
• Seasonal patterns repeat every year.
• Residual values represent irregular fluctuations not explained by trend or seasonality.
• Seasonal effects are strongest near the end of each year.

```python
#ADF Test

result = adfuller(monthly_sales["Sales"])

print("ADF Statistic :", result[0])
print("p-value :", result[1])
```

Stationarity means that the statistical properties of a time series remain constant over time.

If p-value < 0.05:
The series is stationary.

If p-value > 0.05:
The series is non-stationary and needs differencing.

```python
#Differencing

monthly_sales_diff = monthly_sales["Sales"].diff().dropna()

result = adfuller(monthly_sales_diff)

print("ADF Statistic :", result[0])
print("p-value :", result[1])
```

Observation:

After applying first-order differencing, the p-value decreases, indicating that the transformed series becomes stationary and is suitable for forecasting models like SARIMA.

```python
#Plot Differenced Series

plt.figure(figsize=(15,6))

plt.plot(monthly_sales_diff)

plt.title("Differenced Monthly Sales")

plt.grid(True)

plt.savefig(
    "charts/differenced_sales.png",
    bbox_inches="tight"
)

plt.show()
```

Task 2 Observations

Time Series Decomposition Observations:
The overall sales trend shows a gradual increase over the four-year period, indicating that the business experienced steady growth.
A clear seasonal pattern is visible, with sales generally increasing toward the end of each year, suggesting higher customer demand during festive and holiday seasons.
The residual component contains irregular fluctuations that cannot be explained by trend or seasonality alone, indicating occasional unexpected events affecting sales.
The decomposition confirms that both long-term growth and seasonal effects are important components of the sales data.


ADF (Augmented Dickey-Fuller) Test Explanation

What is Stationarity?
A stationary time series has statistical properties such as mean and variance that remain constant over time. Forecasting models like SARIMA generally perform better when the data is stationary.

If p-value > 0.05:
The ADF test produced a p-value greater than 0.05, indicating that the original sales series is non-stationary. Therefore, differencing was applied to remove the trend and stabilize the series before forecasting.

If p-value < 0.05:
The ADF test produced a p-value less than 0.05, indicating that the sales series is already stationary. Therefore, additional differencing is not required.

After Differencing
The differenced series fluctuates around zero without a noticeable upward or downward trend.
Most of the trend has been removed, making the data more suitable for forecasting models.
The variance appears relatively stable over time, indicating improved stationarity.
The differenced series satisfies the assumptions required for statistical forecasting techniques such as SARIMA.

Task 3 – Sales Forecasting Using 3 Different Models

Model 1 — SARIMA (Statistical Model)


```python
#Import Libraries

from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

```python
#Split the Data

train = monthly_sales.iloc[:-3]
test = monthly_sales.iloc[-3:]

print("Training months:", len(train))
print("Testing months:", len(test))
```

```python
#Train the SARIMA Model

sarima_model = SARIMAX(
    train["Sales"],
    order=(1,1,1),
    seasonal_order=(1,1,1,12)
)

sarima_result = sarima_model.fit()

print(sarima_result.summary())
```

Using (1,1,1) and (1,1,1,12) as a reasonable starting configuration for monthly data with yearly seasonality.

```python
#Forecast the Next 3 Months

forecast = sarima_result.get_forecast(steps=3)

forecast_mean = forecast.predicted_mean

confidence = forecast.conf_int()

print(forecast_mean)
```

```python
#Plot Actual vs Forecast

import os

os.makedirs("charts", exist_ok=True)

plt.figure(figsize=(14,6))

plt.plot(
    train.index,
    train["Sales"],
    label="Training Data"
)

plt.plot(
    test.index,
    test["Sales"],
    label="Actual Sales",
    marker="o"
)

plt.plot(
    test.index,
    forecast_mean,
    label="SARIMA Forecast",
    marker="o"
)

plt.fill_between(
    test.index,
    confidence.iloc[:,0],
    confidence.iloc[:,1],
    color="gray",
    alpha=0.3,
    label="Confidence Interval"
)

plt.title("SARIMA Forecast vs Actual Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()

plt.savefig(
    "charts/sarima_forecast.png",
    bbox_inches="tight"
)

plt.show()
```

```python
#Calculate Evaluation Metrics

mae = mean_absolute_error(test["Sales"], forecast_mean)

rmse = np.sqrt(
    mean_squared_error(test["Sales"], forecast_mean)
)

mape = np.mean(
    np.abs((test["Sales"] - forecast_mean) / test["Sales"])
) * 100

print("MAE :", round(mae,2))
print("RMSE:", round(rmse,2))
print("MAPE:", round(mape,2))
```

SARIMA Model Observations

- The SARIMA model captures both trend and seasonal patterns in the monthly sales data.
- A 3-month forecast was generated along with confidence intervals.
- Forecast accuracy was evaluated using MAE, RMSE, and MAPE.
- These metrics will later be compared with Prophet and XGBoost to determine the best forecasting model.

Model 3 — XGBoost for Time Series (ML-based Approach)


```python
#Import Libraries

from xgboost import XGBRegressor

from sklearn.metrics import mean_absolute_error, mean_squared_error

import numpy as np

import pandas as pd
```

```python
#Create Lag Features

xgb_df = monthly_sales.copy()

xgb_df["Lag_1"] = xgb_df["Sales"].shift(1)

xgb_df["Lag_2"] = xgb_df["Sales"].shift(2)

xgb_df["Lag_3"] = xgb_df["Sales"].shift(3)

xgb_df["Rolling_Mean_3"] = (
    xgb_df["Sales"]
    .rolling(3)
    .mean()
)

xgb_df = xgb_df.dropna()

xgb_df.head()
```

```python
#Add Time Features

xgb_df["Month"] = xgb_df["Order Date"].dt.month

xgb_df["Quarter"] = xgb_df["Order Date"].dt.quarter

xgb_df["Year"] = xgb_df["Order Date"].dt.year

xgb_df.head()
```

```python
#Prepare Features

features = [
    "Lag_1",
    "Lag_2",
    "Lag_3",
    "Rolling_Mean_3",
    "Month",
    "Quarter",
    "Year"
]

X = xgb_df[features]

y = xgb_df["Sales"]
```

```python
#Train/Test Split

X_train = X.iloc[:-3]

X_test = X.iloc[-3:]

y_train = y.iloc[:-3]

y_test = y.iloc[-3:]

print(X_train.shape)

print(X_test.shape)
```

```python
#Train XGBoost

xgb_model = XGBRegressor(

    n_estimators=200,

    learning_rate=0.05,

    max_depth=3,

    random_state=42
)

xgb_model.fit(
    X_train,
    y_train
)
```

```python
#Predict

xgb_forecast = xgb_model.predict(
    X_test
)

xgb_forecast
```

```python
#Calculate Metrics

xgb_mae = mean_absolute_error(
    y_test,
    xgb_forecast
)

xgb_rmse = np.sqrt(
    mean_squared_error(
        y_test,
        xgb_forecast
    )
)

xgb_mape = np.mean(

    np.abs(

        (y_test - xgb_forecast)

        / y_test

    )

) * 100

print("MAE :", round(xgb_mae,2))

print("RMSE:", round(xgb_rmse,2))

print("MAPE:", round(xgb_mape,2))
```

```python
#Plot Actual vs Predicted

import os

os.makedirs("charts", exist_ok=True)

plt.figure(figsize=(12,6))

plt.plot(

    y_test.index,

    y_test,

    marker="o",

    label="Actual"

)

plt.plot(

    y_test.index,

    xgb_forecast,

    marker="o",

    label="Predicted"

)

plt.title("XGBoost Forecast vs Actual")

plt.xlabel("Test Months")

plt.ylabel("Sales")

plt.legend()

plt.grid(True)

plt.savefig(

    "charts/xgboost_forecast.png",

    bbox_inches="tight"

)

plt.show()
```

```python
#Feature Importance

importance = pd.DataFrame({

    "Feature": features,

    "Importance": xgb_model.feature_importances_

})

importance = importance.sort_values(

    by="Importance",

    ascending=False

)

importance
```

Task 5 — Anomaly Detection in Sales Data


```python
import pandas as pd

sales_df = pd.read_csv(
    "train excel/train.csv",
    parse_dates=["Order Date", "Ship Date"],
    dayfirst=True
)
```

```python
#Prepare Weekly Sales

weekly_sales = sales_df.groupby(
    pd.Grouper(key="Order Date", freq="W")
)["Sales"].sum().reset_index()

weekly_sales.head()
```

```python
#Isolation Forest

from sklearn.ensemble import IsolationForest

iso = IsolationForest(
    contamination=0.05,
    random_state=42
)

weekly_sales["Anomaly"] = iso.fit_predict(
    weekly_sales[["Sales"]]
)

weekly_sales.head()
```

```python
import matplotlib.pyplot as plt
```

```python
#Plot Isolation Forest Anomalies

import os
os.makedirs("charts", exist_ok=True)
plt.figure(figsize=(15,6))

plt.plot(
    weekly_sales["Order Date"],
    weekly_sales["Sales"],
    label="Weekly Sales"
)

plt.scatter(
    weekly_sales.loc[
        weekly_sales["Anomaly"]==-1,
        "Order Date"
    ],
    weekly_sales.loc[
        weekly_sales["Anomaly"]==-1,
        "Sales"
    ],
    color="red",
    s=80,
    label="Anomaly"
)

plt.title("Isolation Forest Anomaly Detection")

plt.xlabel("Date")

plt.ylabel("Weekly Sales")

plt.legend()

plt.grid(True)

plt.savefig(
    "charts/isolation_forest_anomalies.png",
    bbox_inches="tight"
)

plt.show()
```

```python
#Show Detected Anomalies

weekly_sales[
    weekly_sales["Anomaly"]==-1
]
```

```python
#Z-Score Detection

from scipy.stats import zscore

weekly_sales["ZScore"] = zscore(
    weekly_sales["Sales"]
)

weekly_sales["Z_Anomaly"] = (
    abs(weekly_sales["ZScore"]) > 2
)

weekly_sales.head()
```

```python
#Plot Z-Score Anomalies

plt.figure(figsize=(15,6))

plt.plot(
    weekly_sales["Order Date"],
    weekly_sales["Sales"],
    label="Weekly Sales"
)

plt.scatter(
    weekly_sales.loc[
        weekly_sales["Z_Anomaly"],
        "Order Date"
    ],
    weekly_sales.loc[
        weekly_sales["Z_Anomaly"],
        "Sales"
    ],
    color="green",
    s=80,
    label="Z-Score Anomaly"
)

plt.title("Z-Score Based Anomaly Detection")

plt.xlabel("Date")

plt.ylabel("Weekly Sales")

plt.legend()

plt.grid(True)

plt.savefig(
    "charts/zscore_anomalies.png",
    bbox_inches="tight"
)

plt.show()
```

```python
#Compare Both Methods

comparison = pd.DataFrame({

    "Isolation Forest":

    weekly_sales["Anomaly"]==-1,

    "Z-Score":

    weekly_sales["Z_Anomaly"]

})

comparison.head()
```

```python
#Count Anomalies

print("Isolation Forest Anomalies :",
      (weekly_sales["Anomaly"]==-1).sum())

print("Z-Score Anomalies :",
      weekly_sales["Z_Anomaly"].sum())
```

Observations

- Isolation Forest successfully identified unusual spikes and drops in weekly sales.
- Z-Score detected weeks where sales deviated significantly from the average.
- Some anomalies were identified by both methods, while others were unique to one method.
- This difference indicates that Isolation Forest can detect more complex patterns, whereas Z-Score focuses on statistical deviations.
- The detected anomalies may correspond to festive seasons, promotional campaigns, stock shortages, or unexpected market events.

Task 6 — Product Demand Segmentation using Clustering

```python
#Aggregate Product Data

product_data = sales_df.groupby("Sub-Category").agg(
    Total_Sales=("Sales", "sum"),
    Avg_Order_Value=("Sales", "mean"),
    Sales_Volatility=("Sales", "std"),
    Number_of_Orders=("Sales", "count")
).reset_index()

product_data.head()
```

```python
#Feature Scaling

from sklearn.preprocessing import StandardScaler

features = product_data[
    [
        "Total_Sales",
        "Avg_Order_Value",
        "Sales_Volatility",
        "Number_of_Orders"
    ]
]

scaler = StandardScaler()

scaled_features = scaler.fit_transform(features)
```

```python
#Elbow Method

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import os

os.makedirs("charts", exist_ok=True)

wcss = []

for i in range(1,11):

    kmeans = KMeans(
        n_clusters=i,
        random_state=42,
        n_init=10
    )

    kmeans.fit(scaled_features)

    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))

plt.plot(range(1,11), wcss, marker="o")

plt.title("Elbow Method")

plt.xlabel("Number of Clusters")

plt.ylabel("WCSS")

plt.grid(True)

plt.savefig(
    "charts/elbow_method.png",
    bbox_inches="tight"
)

plt.show()
```

```python
#Apply K-Means

kmeans = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=10
)

product_data["Cluster"] = kmeans.fit_predict(
    scaled_features
)

product_data.head()
```

```python
#PCA Visualization

from sklearn.decomposition import PCA

pca = PCA(n_components=2)

pca_features = pca.fit_transform(scaled_features)

product_data["PCA1"] = pca_features[:,0]

product_data["PCA2"] = pca_features[:,1]
```

```python
#Plot Clusters

plt.figure(figsize=(10,7))

plt.scatter(
    product_data["PCA1"],
    product_data["PCA2"],
    c=product_data["Cluster"],
    s=150
)

for i in range(len(product_data)):

    plt.text(
        product_data["PCA1"][i],
        product_data["PCA2"][i],
        product_data["Sub-Category"][i],
        fontsize=8
    )

plt.title("Product Demand Segments")

plt.xlabel("PCA Component 1")

plt.ylabel("PCA Component 2")

plt.grid(True)

plt.savefig(
    "charts/product_clusters.png",
    bbox_inches="tight"
)

plt.show()
```

```python
#Assign Cluster Labels

cluster_labels = {
    0: "High Volume, Stable Demand",
    1: "Growing Demand",
    2: "Low Volume, High Volatility",
    3: "Declining Demand"
}

product_data["Demand Segment"] = product_data["Cluster"].map(cluster_labels)

product_data[
    ["Sub-Category", "Demand Segment"]
]
```

```python
#Save Results

product_data.to_csv(
    "product_segments.csv",
    index=False
)

product_data.head()
```
